```python
#based on this data, what are the top 10 popular songs over the years
Ts_popular = Ts.sort_values(by='popularity',ascending=False)
print('The most popular songs over the years:')
Ts_popular.head(10)
```

```python
Ts_dance = Ts.sort_values(by='danceability',ascending=False)
print('The most danceable songs over the years:')
Ts_dance.head(10)
```

```python
Tsa=Ts_dance.head(5)
```

```python
#the most danceable songs 
dance = Tsa['danceability']
song = Tsa['song']

# Plotting the histogram for column1
plt.hist(dance, bins=10, alpha=0.5, label='danceability')

# Plotting the histogram for column2
plt.hist(song, bins=10, alpha=0.5, label='song')

plt.xlabel('songs')
plt.ylabel('danceability')
plt.title('The most danceable songs over the years')
plt.xticks(rotation =45)
plt.legend()
plt.show()
```

```python
#To know what year had the most songs
song_data = Ts['year']

# Calculate the mode(s)
mode_result = song_data.mode()

# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(song_data, bins=20, kde=True)  # Adjust the number of bins as needed
plt.title('The year with the highest song realease')
plt.xlabel('year')
plt.ylabel('Frequency')

# Highlight mode(s)
for mode in mode_result:
    plt.axvline(x=mode, color='red', linestyle='--', label=f'Mode: {mode}')
plt.legend()

plt.show()
```

```python
most_common_value = Ts['genre'].value_counts().idxmax()

print("The most frequently occurring string in the column is:", most_common_value)
```

```python
value_counts = Ts['genre'].value_counts()

# Get the most common value
most_common_value = value_counts.idxmax()

# Plotting boxplot
plt.figure(figsize=(8, 6))
value_counts.plot(kind='box')
plt.title('The most popular genre of music in the dataset')
plt.ylabel('genre')
plt.axhline(value_counts[most_common_value], color='r', linestyle='--',label='Most Common Value: {}'.format(most_common_value))
plt.legend()
plt.show()
```

```python
correlation_matrix = Ts[['danceability', 'popularity']].corr()

# Plotting correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Plot between the popularity and danceability of the songs')
plt.show()
```

```python
value_counts = Ts['artist'].value_counts()

# Get the most common value
most_common_value = value_counts.idxmax()

# Plotting boxplot
plt.figure(figsize=(8, 6))
value_counts.plot(kind='box')
plt.title('The artist that has the most songs')
plt.ylabel('Frequency')
plt.axhline(value_counts[most_common_value], color='r', linestyle='--', label='Most Common Value: {}'.format(most_common_value))
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
popularity_by_year = Ts.groupby('year')['popularity'].mean()

# Assuming 'popularity_by_year' is a pandas Series with years as index
# Convert the index to datetime for time series analysis
popularity_by_year.index = pd.to_datetime(popularity_by_year.index, format='%Y')

# Split the data into training and testing sets
train_data = popularity_by_year.iloc[:-2]  
test_data = popularity_by_year.iloc[-2:]  

# Fit ARIMA model
model = ARIMA(train_data, order=(1,1,1))  
model_fit = model.fit()

# Forecast future popularity
future_years = pd.date_range(start=popularity_by_year.index[-1], periods=5, freq='Y')
forecast = model_fit.forecast(steps=5)  # Forecast for the next 5 years

# Plotting
plt.plot(popularity_by_year.index, popularity_by_year, label='Actual Popularity')
plt.plot(future_years, forecast, label='Forecasted Popularity')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('Forecasting Popularity of Songs')
plt.legend()
plt.show()
```
